# Programa de Cadastro de SÃ©ries de TV

Este programa permite que o usuÃ¡rio cadastre as sÃ©ries de TV que estÃ¡ assistindo, juntamente com a temporada atual.
AlÃ©m disso, o usuÃ¡rio pode adicionar notas crÃ­ticas para as sÃ©ries cadastradas.
O programa interage com o usuÃ¡rio, fornecendo um menu de opÃ§Ãµes para adicionar novas sÃ©ries, adicionar notas crÃ­ticas, visualizar todas as sÃ©ries cadastradas ou sair do programa.

## Funcionalidades:
1. Cadastro de sÃ©ries e temporadas.
2. AdiÃ§Ã£o de notas crÃ­ticas para sÃ©ries cadastradas.
3. VisualizaÃ§Ã£o de todas as sÃ©ries cadastradas.
4. SaÃ­da do programa.

## InstruÃ§Ãµes:
1. Ao iniciar o programa, serÃ¡ solicitado o nome, idade e paÃ­s do usuÃ¡rio, com uma mensagem de boas-vindas.
2. O usuÃ¡rio pode escolher adicionar uma nova sÃ©rie, adicionar uma nota crÃ­tica, visualizar todas as sÃ©ries ou sair do programa.


```python
def boas_vindas():
    nome = input("Qual Ã© o seu nome? ")
    idade = input("Qual Ã© a sua idade? ")
    pais = input("Qual Ã© o seu paÃ­s? ")
    print(f"\nBem-vindo(a), {nome} de {idade} anos, do {pais}!\n")
    return nome

def adicionar_serie(series):
    serie = input("Qual sÃ©rie vocÃª estÃ¡ assistindo? ")
    temporada = input("Qual temporada vocÃª estÃ¡ atualmente? ")
    series[serie] = {"temporada": temporada, "nota_critica": ""}
    print(f"SÃ©rie '{serie}' adicionada com sucesso!\n")

def adicionar_nota_critica(series):
    serie = input("Para qual sÃ©rie vocÃª quer adicionar uma nota crÃ­tica? ")
    if serie in series:
        nota_critica = input("Digite sua nota crÃ­tica para essa sÃ©rie: ")
        series[serie]["nota_critica"] = nota_critica
        print(f"Nota crÃ­tica para a sÃ©rie '{serie}' adicionada com sucesso!\n")
    else:
        print(f"SÃ©rie '{serie}' nÃ£o encontrada no cadastro.\n")

def mostrar_series(series):
    if not series:
        print("Nenhuma sÃ©rie cadastrada.\n")
    else:
        for serie, detalhes in series.items():
            print(f"SÃ©rie: {serie}, Temporada: {detalhes['temporada']}, Nota CrÃ­tica: {detalhes['nota_critica']}\n")

def main():
    series = {}
    nome = boas_vindas()

    while True:
        print("O que vocÃª gostaria de fazer?")
        print("1. Adicionar uma nova sÃ©rie")
        print("2. Adicionar uma nota crÃ­tica para uma sÃ©rie")
        print("3. Mostrar todas as sÃ©ries cadastradas")
        print("4. Sair do programa")

        escolha = input("Digite o nÃºmero da sua escolha: ")

        if escolha == "1":
            adicionar_serie(series)
        elif escolha == "2":
            adicionar_nota_critica(series)
        elif escolha == "3":
            mostrar_series(series)
        elif escolha == "4":
            print(f"AtÃ© logo, {nome}!")
            break
        else:
            print("Escolha invÃ¡lida. Tente novamente.\n")

if __name__ == "__main__":
    main()
```
